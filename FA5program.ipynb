```python
price = int(input ("How much is the object? "))
quantity = int(input ("How many of the object is there? " ))

def total(result):
  print (f"The total is {result}")

def multiply(price, quantity):
  return price * quantity

result1 = multiply(price, quantity)

total(result1)
```
